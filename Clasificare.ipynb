```python
from sklearn import tree, datasets, metrics
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, tree, svm, neural_network

from sklearn import ensemble
```

```python
date = []
etichete = []
```

## Citirea bazei de date

```python
with open('breast-cancer-wisconsin.data', 'r') as f:
    for line in f:
        list_of_data = line.split(',')
        list_of_data[-1]=list_of_data[-1].replace('\n', '')
        list_of_data = [int(i) if i != '?' else np.nan for i in list_of_data]
        # print(list_of_data)
        etichete.append(list_of_data.pop())
        date.append(list_of_data[1:])
```

## Inlocuirea valorilor lipsa cu media

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='most_frequent')
date = my_imputer.fit_transform(date)
```

## Impartirea datelor

```python
date_train = date[:524, :]
date_test = date[524:, :]

etichete_train = etichete[:524]
etichete_test = etichete[524:]
```

## Random Forest

```python
clf = ensemble.RandomForestClassifier()

clf.fit(date_train,etichete_train)
```

```python
predictii = clf.predict(date_test)

acuratete = metrics.accuracy_score(y_true = etichete_test, y_pred = predictii)
print('Acuratete test = ',acuratete)
```

## SVM

```python
clf = svm.SVC(kernel='rbf', C=1, gamma='auto')

clf.fit(date_train,etichete_train)
```

```python
predictii = clf.predict(date_test)

acuratete = metrics.accuracy_score(y_true = etichete_test, y_pred = predictii)
print('Acuratete test = ',acuratete)
```

## MLP

```python
clf = neural_network.MLPClassifier(learning_rate_init=0.1)
clf.fit(date_train,etichete_train)
```

```python
predictii = clf.predict(date_test)


acuratete = metrics.accuracy_score(y_true = etichete_test, y_pred = predictii)
print('Acuratete test = ',acuratete)
```
