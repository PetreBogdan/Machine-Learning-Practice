## Importarea datelor

```python
from sklearn import tree, datasets, metrics
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, tree, svm, neural_network

from sklearn import ensemble
```

```python
date = []
etichete = []
```

## Citirea bazei de date

```python
with open('auto-mpg.data','r') as f:
    for line in f:
        list_of_data = line.split()
        # print(list_of_data)
        list_of_data = [float(i) if i != '?' else np.nan for i in list_of_data[0:8]]
        date.append(list_of_data[1:8])
        etichete.append(list_of_data[0])
```

## Inlocuirea valorilor lipsa

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='most_frequent')
date = my_imputer.fit_transform(date)
```

## Impartirea datelor

```python
date_train = date[:298, :]
date_test = date[298:, :]

etichete_train = etichete[:298]
etichete_test = etichete[298:]

print(len(date_train), len(etichete_train))
print(len(date_test), len(etichete_test))
```

## Arbori de decizie si Random Forest

```python
regr = tree.DecisionTreeRegressor(criterion='absolute_error', max_depth=None)
regr.fit(date_train,etichete_train)
```

```python
regr = ensemble.RandomForestRegressor(criterion='absolute_error', max_depth=None)
regr.fit(date_train, etichete_train)
```

```python
predictii = regr.predict(date_test)
Err = metrics.mean_squared_error(y_true=etichete_test, y_pred=predictii)
print('Eroarea patratica medie =', Err)
```

## SVM

```python
regr = svm.SVR(kernel='linear', gamma='auto', C=0.1)
regr.fit(date_train, etichete_train)
```

```python
predictii = regr.predict(date_test)
Err = metrics.mean_squared_error(y_true=etichete_test, y_pred=predictii)
print('Eroarea patratica medie =', Err)
```

## MLP

```python
regr = neural_network.MLPRegressor(learning_rate_init=0.01, hidden_layer_sizes=(200,))
regr.fit(date_train,etichete_train)
```

```python
predictii = regr.predict(date_test)
Err = metrics.mean_squared_error(y_true=etichete_test, y_pred=predictii)
print('Eroarea patratica medie =', Err)
```
